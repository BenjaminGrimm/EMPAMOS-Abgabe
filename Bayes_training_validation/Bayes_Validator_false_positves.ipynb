### Stopwörter entfernt und keine lemmatisierung

```python
directoryname = "Bayes_no_lemmatize_stopword_removed"
os.chdir(directoryname)
client = empamos_rest_client()

```

```python
pattern_ids = [pattern['pattern_id'] for pattern in client.get_patterns()]
for id in pattern_ids:
    factory = TrainingSetFactory(4)
    df = factory.trainings_dataframe(id,False,True,False)
    counter = 0      
    counterWrong = 0
    #if model exists
    if(path.isfile("Bayes-"+str(id)+".sav")):
        with open(str(id)+"Bayes-.txt","w") as file:
            print("\n")
            file.write(str(len(df)))
            file.write("\n")
            model = pickle.load(open("Bayes-"+str(id)+".sav", 'rb'))
            count_vector = pickle.load(open("Counter-"+str(id)+".sav", 'rb'))
            for i in range(0,len(df)):
                sentence = count_vector.transform([df.iloc[i][0]])
                prediction = model.predict(sentence)
                if(prediction[0] == 1):
                    file.write(df.iloc[i][0])
                    file.write("\n")
                    counterWrong+=1                
                counter+=1
    
        print(str(id)+" Richtig zugeordnet: " + str(counter - counterWrong))
        print(str(id)+" In Prozent: " + str(100 * (counter - counterWrong) / counter) + "%")
        print(str(id)+" Falsch zugeordnet: " + str(counterWrong))
        print(str(id)+" In Prozent: " + str(100 * counterWrong / counter) + "%")
    else:
        print("No Model found!"+str(id))
```
