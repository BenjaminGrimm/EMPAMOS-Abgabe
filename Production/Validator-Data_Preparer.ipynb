```python

def prepare_pattern(pattern_id):
    print("Preparing pattern " + str(pattern_id))
    directory = "Validator_sentences"
    os.makedirs(directory, exist_ok=True)
    pos_file_name = str(pattern)+"-pos.txt"
    neg_file_name = str(pattern)+"-neg.txt"
    min_sentence_length = 4
    
    pos_sents = client.get_pattern_evidence_sentences(pattern_id)
    with open(directory + "/" + pos_file_name, "w") as f: 
        for sent in pos_sents:
            f.write("%s\n" % sent)
        
    games_negative_patterns = client.evidence_pattern_search('','',pattern_id,'')
    
    #There is always one entry with descriptive information
    #If there is nothing else we won't find a fitting game
    if(len(games_negative_patterns) == 1):
        print("No negative pattern found")
        return
    print("Negative pattern found. Preparing...")
    #Get rid of the descriptive entry
    del games_negative_patterns[0]
    neg_sents = []
    for evidence in games_negative_patterns:
        result = session.execute("select text from working_text where SourceUID = '" + str(evidence["solr_id"]) + "';")
        text = ""
        for line in result:
            text+=line.text
        if len(text.strip()) == 0:
            continue
        nlp_text = nlp(text)
        for sent in nlp_text.sents:
            if(len(sent) >= min_sentence_length):
                neg_sents.append(str(sent))
    if(len(neg_sents) > len(pos_sents)):
        neg_sents = random.sample(neg_sents,len(pos_sents))  
            
    with open(directory + "/" + neg_file_name, "w") as f: 
        for sent in neg_sents:
            f.write("%s\n" % sent)
```
