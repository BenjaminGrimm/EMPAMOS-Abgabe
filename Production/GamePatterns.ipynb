Klasse game_patterns
Enthält Informationen darüber welche Patterns an welchen Stellen einer Anleitung vorkommen.
Die Patterns können beispielsweise vom pattern_localizer gefunden und anschließend in die game_patterns Klasse eingetragen werden. Lässt sich als JSON serialisieren und deserialisieren.

    id              -Spiel ID
    sentences       -Anleitungstext als Liste von Sätzen
    patterns        -Dictionary mit Patterns die in der Anleitung vorkommen. 
                     Schema: Key = Pattern ID, Value = Liste von Sätzen die das Pattern enthalten.
                     
Funktionen (ohne Getter und Setter)
    
    get_sentence_count
        returns      -Anzahl der Sätze in der Anleitung
        
    add_pattern
        patternID    -Id des Patterns
        sentenceList -List von Satz-Nummern mit Vorkommen des Patterns
    
        
    get_sentences_with_patterns
        names        -Dictionary mit Namen der Patterns. Struktur: Key=PatternID Value=Name
        returns      -Alle Sätze in annotierter Form durch Zeilenumbrüche getrennt als String
        
    print_sentences_with_patterns  
    
        wie get_sentences_with_patterns aber als Ausgabe


    print_verbose    
        names        -Dictionary mit Namen der Patterns. Struktur: Key=PatternID Value=Name
        Ausgabe aller relevanter Informationen des game_patterns Objekts
        
    toJSON
        Conversion als JSON. Exportierte JSONs können wiederhergestellt werden.
        


```python
import json
```

```python
class game_patterns:
        
    def __init__ (self, *args, **kwargs):
        if "id" in kwargs:
            self.id = kwargs["id"]
        else:
            self.id = -1
        if "sentences" in kwargs:
            self.sentences = kwargs["sentences"]
        else:
            self.sentences = []
        if "patterns" in kwargs:
            self.patterns = kwargs["patterns"]
            self.patterns = {int(k):v for k,v in self.patterns.items()}
        else:
            self.patterns = {}
          
        
    def set_ID (self,id):
        self.id = id
        
    def get_ID (self):
        return self.id
        
    def set_sentences (self, sentences):
        self.sentences = sentences
        
    def get_sentences (self):
        return self.sentences
        
    def get_sentence_count (self):
        return len(self.sentences)
        
    def add_pattern (self, patternID, sentenceList):
        self.patterns[patternID] = sentenceList
        
    def get_pattern (self, patternID):
        return self.patterns[patternID]
    
    def get_sentences_patterns(self):
        sent_pats=[] 
        for a in range(0,len(self.sentences)):
            sent_pats.append([])
        for pat_id, sent_list in self.patterns.items():
            for sent in sent_list:
                sent_pats[sent].append(pat_id)
        return sent_pats
    
    def print_sentences_with_patterns_no_names(self):
        sent_pats = self.get_sentences_patterns()
        for i in range(0,len(self.sentences)):
            print(str(i)+": "+self.sentences[i] + "; " + str(sent_pats[i]) )
            
    
    def print_sentences_with_patterns(self,names):
        sent_pats = self.get_sentences_patterns()
        for i in range(0,len(self.sentences)):
            pats = "["
            for sp in sent_pats[i]:
                pats = pats + str(sp)
                if sp in names:
                    pats = pats + ": " +names[sp] 
                pats = pats + ", "
            pats = pats + "]"
            print(str(i)+": "+self.sentences[i] + "; " + pats )
            
    def get_sentences_with_patterns(self,names):
        sent_pats = self.get_sentences_patterns()
        output_list = []
        for i in range(0,len(self.sentences)):
            pats = "["
            for sp in sent_pats[i]:
                pats = pats + str(sp)
                if sp in names:
                    pats = pats + ": " +names[sp] 
                pats = pats + ", "
            pats = pats + "]"
            output_list.append(str(i)+": "+self.sentences[i] + "; " + pats)
        return "\n".join(output_list)
    
    def print_verbose(self, names):
        print("Patterns of game " + str(self.get_ID()))
        print("Sentence Count " + str(self.get_sentence_count()))
        print("Total number of patterns: " +str(len(self.patterns)))
        print(self.patterns)
        print("")
        self.print_sentences_with_patterns(names)
        print("")
        
    def toJSON(self):
        return json.dumps(self, default=lambda o: o.__dict__,
            sort_keys=False, indent=2)
```
