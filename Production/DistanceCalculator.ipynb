Distance Calculator
Bietet Funktionen zur Bestimmung von Abständen von Patterns in einer Anleitung

    calculate_distance(patterns, fn_dist)
        patterns:     -Patterns-Dictionary. Benötigt Struktur Key = Pattern_ID, Value = Liste von Vorkommen. 
                   (Entspricht patterns in der Klasse game_patterns)
        fn_dist:      -Gewählte Abstandsfunktion
  
Verfügbare Abstandsfuktionen:
<ul>
    <li>distance_minimum:      Minimaler paarweiser Abstand zweier Elemente</li>
<li>distance_average:      Durchschnittlicher minimaler paarweiser Abstand</li>
<li>distance_hausdorff:    Hausdorff-Metrik</li>
<li>distance_cuddle:       "Ankuschelabstand"</li>
</ul>


def calculate_distance(patterns, fn_dist):
    distances = {}
    keys = list(patterns.keys())
    for i in range(0,len(keys)-1):
        for j in range (i+1,len(keys)):
            distances[str(keys[i])+"-"+str(keys[j])] = fn_dist(patterns[keys[i]],patterns[keys[j]])
    return distances


                                                                                         

```python
def calculate_distance(patterns, fn_dist):
    distances = {}
    keys = list(patterns.keys())
    for k in keys:
        distances[k] = {}
        
    for i in range(0,len(keys)):
        distances[keys[i]][keys[i]] = 0
        for j in range (i+1,len(keys)):
            d = fn_dist(patterns[keys[i]],patterns[keys[j]])
            distances[keys[i]][keys[j]] = d
            distances[keys[j]][keys[i]] = d    
    return distances
```

```python
def distance_minimum(list1, list2):
    d = 1234567890         #Einfach eine große Zahl. Sicher größer als die Anzahl der Sätze in der Anleitung
    for i in list1:
        for j in list2:
            d = min(d,abs(i-j))
    return d
       
def distance_average(list1, list2):
    d = 0
    for i in list1:
        part = 1234567890   #Einfach eine große Zahl. Sicher größer als die Anzahl der Sätze in der Anleitung
        for j in list2:
            #suche den nähsten Punkt in der anderen Liste
            part = min(part,abs(i-j))
        d += part
    #Nochmal mit getauschten Rollen
    for i in list2:
        part = 1234567890
        for j in list1:
            part = min(part,abs(i-j))
        d += part
    return (d/(len(list1)+len(list2))) 

def distance_hausdorff(list1, list2):
    h1 = 0
    for i in list1:
        d = 1234567890
        for j in list2:
            d = min(d,abs(i-j))
        h1 = max(h1,d)
    #Nochmal mit getauschten Rollen
    h2 = 0
    for i in list2:
        d = 1234567890
        for j in list1:
            d = min(d,abs(i-j))
        h2 = max(h2,d)
    return max(h1,h2) 

def distance_cuddle(list1, list2):
    d1 = 0
    for i in list1:
        part = 1234567890
        for j in list2:
            part = min(part,abs(i-j))
        d1 += part

    #Nochmal mit getauschten Rollen
    d2 = 0
    for i in list2:
        part = 1234567890
        for j in list1:
            part = min(part,abs(i-j))
        d2 += part
        
    return min(d1/len(list1), d2/len(list2))

```
